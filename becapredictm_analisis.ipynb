```python
# importamos librerias necesarias
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
import pandas as pd 
```

```python
# importando dataset
becariosp = pd.read_csv('perdidadebecas.csv')
print(becariosp)
```

```python
# Imprimiendo las 5 primeras instancias seleccionando algunas caracteristicas
print(becariosp.head()[['region', 'tipoinstitucion', 'carreradescripcion','genero', 'motivo','convocatoria']])
```

```python
# tabla de frecuencia de motivo de perdida de beca o conteo por motivo de perdida de beca
pd.value_counts(becariosp['motivo'])
```

```python
# tabla de frecuencia relativa del motivo de perdida de beca de los becarios
100 * becariosp['motivo'].value_counts() / len(becariosp['motivo'])
```

```python
# Gráfico de barras del motivo de perdida del becario
plot = becariosp['motivo'].value_counts().plot(kind='barh', title='Motivo de perdida de beca')
```

```python
# Tabla de contingencia motvio / region
pd.crosstab(index=becariosp['motivo'],columns=becariosp['region'], margins=True)  # margins_name="TOTAL"
```

```python
# tabla de contingencia en porcentajes relativos total region/motivo
pd.crosstab(index=becariosp['motivo'], columns=becariosp['region'],margins=True).apply(lambda r: r/len(becariosp) *100,axis=1)
```

```python
# tabla de contingencia en porcentajes relativos total motivo/tipoinstitucion
pd.crosstab(index=becariosp['motivo'], columns=becariosp['tipoinstitucion'],margins=True).apply(lambda r: r/len(becariosp) *100,axis=1)
```

```python
# Gráfico de barras de motivo/region 
pd.crosstab(index=becariosp['motivo'], columns=becariosp['tipoinstitucion']).apply(lambda r: r/r.sum() *100,axis=1).plot(kind='bar',legend=True, figsize=(25, 25))
```

```python
# Tabla de contingencia motivo / genero
pd.crosstab(index=becariosp['motivo'],columns=becariosp['genero'], margins=True)  # margins_name="TOTAL"
```

```python
# tabla de contingencia en porcentajes relativos total motivo/genero
pd.crosstab(index=becariosp['motivo'], columns=becariosp['genero'],margins=True).apply(lambda r: r/len(becariosp) *100,axis=1)
```

```python
# Gráfico de barras de motivo/genero 
pd.crosstab(index=becariosp['motivo'], columns=becariosp['genero']).apply(lambda r: r/r.sum() *100,axis=1).plot(kind='bar',legend=True, figsize=(20, 8))
```
